```python
#Gráfico de Linha Simples:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 15, 7]

plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linha Simples')
plt.show()
```

```python
#Gráfico de Barras:

import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C', 'D', 'E']
valores = [8, 12, 9, 15, 10]

plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
plt.show()
```

```python
#Gráfico de Dispersão:

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 15, 7]

plt.scatter(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão')
plt.show()
```

```python
#Gráfico de Pizza:

import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # Explodir a segunda fatia (B)

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Gráfico de Pizza')
plt.show()
```

```python
#Histograma:

import matplotlib.pyplot as plt
import numpy as np

dados = np.random.randn(1000)  # Dados aleatórios

plt.hist(dados, bins=20, edgecolor='k')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')
plt.show()
```

```python
#Gráfico de Caixa (Box Plot):

import matplotlib.pyplot as plt
import numpy as np

dados = np.random.randn(100)

plt.boxplot(dados)
plt.ylabel('Valores')
plt.title('Gráfico de Caixa')
plt.show()
```
