```python
#importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#gerar dados para o modelo
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)
#print(data)

#definir quantos centros ter no modelo
kmeans = KMeans(n_clusters=4)

#treinar o modelo
kmeans.fit(data)

#encontrar pontos centrais de cada grupo de individuos para servir como referencia
centroids = kmeans.cluster_centers_

#obter as atribuações de clusters para cada ponto
labels = kmeans.labels_

#plotar
plt.scatter(data[:,0], data[:,1], c=labels, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:,1], marker='X',s=200 ,color='red')
plt.xlabel('dimensao 1')
plt.ylabel('dimensao 2')
plt.title('resultado kmeans')

plt.show()
```
