```python
import cv2

# Inicialize a captura de vídeo a partir da webcam (0) ou de um arquivo de vídeo
cap = cv2.VideoCapture(0)  # Use 0 para a webcam padrão

# Verifique se a captura de vídeo foi aberta com sucesso
if not cap.isOpened():
    print("Erro ao abrir a captura de vídeo.")
else:
    while True:
        # Captura um quadro do vídeo
        ret, frame = cap.read()
        
        # Verifique se a captura do quadro foi bem-sucedida
        if not ret:
            print("Erro ao capturar o quadro.")
            break

        # Exibe o quadro capturado
        cv2.imshow("Captura de Video", frame)

        # Verifique se a tecla 'q' foi pressionada para sair
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Libere a captura de vídeo e feche a janela
    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2

# Inicialize a captura de vídeo a partir da webcam (0) ou de um arquivo de vídeo
cap = cv2.VideoCapture(0)  # Use 0 para a webcam padrão

# Carregue o classificador Haar Cascade para detecção de rosto
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Verifique se o classificador foi carregado com sucesso
if face_cascade.empty() or not cap.isOpened():
    print("Erro ao carregar a captura de vídeo ou o classificador de rosto.")
else:
    while True:
        # Captura um quadro do vídeo
        ret, frame = cap.read()

        # Verifique se a captura do quadro foi bem-sucedida
        if not ret:
            print("Erro ao capturar o quadro.")
            break

        # Converta o quadro para tons de cinza para a detecção de rosto
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detecte rostos no quadro
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

        # Desenhe retângulos ao redor dos rostos detectados
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Exiba o quadro com a detecção de rosto
        cv2.imshow("Deteccao de Rosto", frame)

        # Verifique se a tecla 'q' foi pressionada para sair
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Libere a captura de vídeo e feche a janela
    cap.release()
    cv2.destroyAllWindows()
```
