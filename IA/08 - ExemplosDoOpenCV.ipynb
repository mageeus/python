```python
'''
Objetivo detecção de rostos
'''
import cv2

# Carregar o classificador de rosto pré-treinado
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Carregar a imagem
imagem = cv2.imread('pessoas2.jpeg')
#imagem = cv2.imread('pessoas2.jpg')

# Converter a imagem para tons de cinza 
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Detectar rostos na imagem
faces = face_cascade.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Desenhar retângulos ao redor dos rostos detectados
for (x, y, w, h) in faces:
    cv2.rectangle(imagem, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Exibir a imagem com os rostos detectados
cv2.imshow('Rostos Detectados', imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
