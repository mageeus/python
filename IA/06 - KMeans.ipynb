```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Gerar dados de exemplo
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)
'''
Eu poderia fazer algo como no unpack? S
_, _ = make_blobs(n_samples=300, centers=4, random_state=42)
'''

#print(data)
print(_)

# Criar um objeto KMeans com 4 clusters
kmeans = KMeans(n_clusters=4)
#kmeans = KMeans()

# Treinar o modelo nos dados
kmeans.fit(data)

# Obter os centróides dos clusters
centroids = kmeans.cluster_centers_
print(centroids)

# Obter as atribuições de clusters para cada ponto de dados
labels = kmeans.labels_

# Plotar os pontos de dados e os centróides
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red')
plt.xlabel('Dimensão 1')
plt.ylabel('Dimensão 2')
plt.title('Resultado do K-Means')
#plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Gerar dados de exemplo simulados
data, _ = make_blobs(n_samples=300, centers=4, n_features=2, random_state=42)

# Lista para armazenar os valores do Silhouette Score
silhouette_scores = []

# Testar diferentes valores de k
for k in range(2, 8):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(data, labels))

# Plotar os valores do Silhouette Score para diferentes valores de k
plt.plot(range(2, 8), silhouette_scores, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Avaliação do K-Means usando Silhouette Score')
plt.show()
```
