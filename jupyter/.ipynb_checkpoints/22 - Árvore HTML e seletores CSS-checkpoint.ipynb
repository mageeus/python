# 73 - rvore HTML e seletores CSS

Os elementos html se organizam em rvore dentro da pgina, onde vrios elementos embutem outros em si
Ex:
HTML <br />
....HEAD <br />
........TITLE <br />
....BODY <br />
........DIV <br />
............A <br />
................IMG <br />
........DIV <br />
............A <br />
        
        
Esse conceito  usado para montar o XPATH, assim como os objetos gerenciados pelo Selenium.
Ento se voc pegar um elemento PAI seus filhos sero includos no objeto criado pelo Selenium.

https://www.w3schools.com/cssref/css_selectors.asp

assim podemos trazer um elemento ou diversos deles, apenas com a lgica de seleo do CSS.
Por exemplo:
	Elementos com o tipo de tag a
