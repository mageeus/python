# 73 - Árvore HTML e seletores CSS

Os elementos html se organizam em árvore dentro da página, onde vários elementos embutem outros em si
Ex:
HTML <br />
....HEAD <br />
........TITLE <br />
....BODY <br />
........DIV <br />
............A <br />
................IMG <br />
........DIV <br />
............A <br />
        
        
Esse conceito é usado para montar o XPATH, assim como os objetos gerenciados pelo Selenium.
Então se você pegar um elemento PAI seus filhos serão incluídos no objeto criado pelo Selenium.

https://www.w3schools.com/cssref/css_selectors.asp

Uma forma útil de selecionar elementos numa página, é ade usarmos os seletores CSS,
assim podemos trazer um elemento ou diversos deles, apenas com a lógica de seleção do CSS.
Por exemplo:
	Elementos com o tipo de tag a
	Elementos com a classe botão
	Elementos com o id principal
